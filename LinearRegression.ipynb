```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv('Area.csv')
print(data)
```

```python
plt.scatter(data.Area,data.Price)
```

```python
regression=linear_model.LinearRegression()
regression.fit(data[['Area']],data.Price)
```

```python
regression.predict([[4500]])
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
a=regression.coef_*4200+regression.intercept_
print(a)
```

```python
regression=linear_model.LinearRegression()
regression.fit(data[['Area','Location']],data.Price) #area and loc dependent, price is independent
```

```python
regression.predict([[3600,5]])
```

```python
regression.predict([[3600,5],[2400,10],[3600,5],[5000,3],[2500,4]])
```
